### Problem 3: 온도변환 프로그램
😄화씨온도와 섭씨온도 간 변환공식은 아래와 같다. 화씨온도 100은 섭씨온도 몇인가? 반대로 섭씨온도 50은 화씨온도 몇인가?
화씨온도를 저장할 변수 f와 섭씨온도를 저장할 변수 c를 만들고
변환공식을 이용하여 코드를 작성하고 결과를 출력해보세요.

변환 공식

<font color='blue'>`화씨(f) → 섭씨(c)`</font>
```python
c = 5.0 / 9.0 * (f - 32.0)
```
<font color='blue'>`섭씨(c) → 화씨(f)`</font>
```python
f = 9.0 / 5.0 * c + 32.0
```
---

Output Sample
```
37.77777777777778
```
Output Sample
```
122.0
```

📜 참고: 코드를 실행하면 화씨온도 100이 섭씨온도 약 37.78이고 섭씨온도 50이 화씨온도 122임을 알 수 있다.
복잡한 연산에서는 괄호의 위치에 특히 유의해야 한다.
