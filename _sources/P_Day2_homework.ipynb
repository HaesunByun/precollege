# 서울대학교 예비대학 디딤돌 프로그램

# Homework-Day2:

### Problem: 합격, 불합격, 과락 판단
😄이름과 국어, 수학, 컴퓨팅기초 점수(정수) 세 개를 입력받아 다음 조건에 따라 결과를 출력하는 프로그램을 작성하시오.

- 합격: 평균이 60점 이상이고 모두 40점 이상
- 과락: 평균이 60점 이상이나 40점 미만 과목이 있음
- 불합격: 평균이 60점 미만

---

<font color='blue'>`Test case-1`</font>

Input Sample
```
홍길동
87
37
77
```
Output Sample
```
이름:
국어:
수학:
컴기:
홍길동님은 과락입니다.
```
<font color='blue'>`Test case-2`</font>

Input Sample
```
홍길동
87
57
77
```
Output Sample
```
이름:
국어:
수학:
컴기:
홍길동님은 합격입니다.
```
<font color='blue'>`Test case-3`</font>

Input Sample
```
홍길동
62
37
68
```
Output Sample
```
이름:
국어:
수학:
컴기:
홍길동님은 불합격입니다.
```
