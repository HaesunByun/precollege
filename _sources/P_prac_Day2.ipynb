### Problem 2: 온도 변환 프로그램

😄 아래 공식을 적용하여 화씨온도와 섭씨온도를 입력 받아 변환하는 온도 변환 프로그램을 작성하세요.

변환 공식

<font color='blue'>`화씨(f) → 섭씨(c)`</font>
```python
c = 5.0 / 9.0 * (f - 32.0)
```
<font color='blue'>`섭씨(c) → 화씨(f)`</font>
```python
f = 9.0 / 5.0 * c + 32.0
```
