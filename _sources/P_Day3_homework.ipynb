# 서울대학교 예비대학 디딤돌 프로그램

# Homework-Day3:

### Problem: "그만"하면 멈추는 똑똑한 앵무새 프로그램
😄조련사의 말을 따라하는 앵무새 프로그램을 작성하시오. 조련사가 입력한 말을 앵무새처럼 따라 말하는 역할을 합니다. 하지만, 조련사가 **"그만"**이라고 입력하면 프로그램이 종료되며, 마지막으로 **"앵무새가 조용해졌습니다."**라는 메시지를 출력해야 합니다.

---

<font color='blue'>`Test case-1`</font>

Input Sample
```
너 이름이 뭐니?
난 조련사야.
잘따라 하는구나.
날씨가 춥구나.
파이썬은 재밌어.
그만
```
Output Sample
```
조련사:
앵무새: 너 이름이 뭐니?
조련사:
앵무새: 난 조련사야.
조련사:
앵무새: 잘따라 하는구나.
조련사:
앵무새: 날씨가 춥구나.
조련사:
앵무새: 파이썬은 재밌어.
앵무새가 조용해졌습니다.
```
