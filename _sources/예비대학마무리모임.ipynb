## Q3, Q4
- print()에서 +와 ,와 f-string의 출력 차이

```
name='홍길동'
age=20
```

- `홍길동님의 나이는 20살입니다.` 이와 같이 정확하게 출력하기를 원했을 때,


- '+'연산자를 사용하는 경우, 변수 age가 정수형이므로 str()로 타입변환을 해줘야 한다.
- `나이는` 다음에 공백까지 생각해줘야 한다.

````python
name='홍길동'
age=20
print(name+'님의 나이는'+str(age)+'살 입니다.')
print(name+'님의 나이는 '+str(age)+'살 입니다.')
````

- 콤마(,)를 사용하는 경우, 콤마 있는 부분이 자동 한칸 띄어쓰기가 되므로 `20살`과 같이 나오게 하려면 sep='' 옵션을 줘서 띄어쓰기를 없애줘야 한다. 
- 이때 올바르게 띄어써준 부분까지 없애지기 때문에 일부러 공백을 넣어줘야 한다.

````python
print(name, '님의 나이는', age, '살 입니다.')
print(name, '님의 나이는', age, '살 입니다.', sep='')
print(name, '님의 나이는 ', age, '살 입니다.', sep='')
````

- f-string으로 처리하는 것이 가장 편하다.
- 출력하고자 하는 형태 그대로 만들고, `변수명`이나 `연산식` 또는 `함수호출` 등이 들어가야 할 자리에 {}를 씌어 넣어주면 된다.

````python
print(f'{name}님의 나이는 {age}살입니다.')
````

## Q8

슬라이싱(Slicing)이란 리스트의 아이템 일부를 자르는 것이다. 형식은 다음과 같다.

```
리스트변수[start:end:step]
```
- start: 시작 인덱스, 디폴트는 step이 양수이면 인덱스 0, step이 음수이면 인덱스 -1이다.
- end: 끝 인덱스+1
- step: 이동간격, 양수이면 오른쪽으로 step만큼 이동한다. 음수이면 왼쪽으로 step만큼 이동한다.

- start, end, step 모두 생략 가능하다.

- 아래의 코드를 통해 실습해보자. 
- 시작 인덱스의 생략으로 처음부터 인덱스 3 이전까지 오른쪽으로 이동하여 슬라이싱한다.

````python
shape= ['turtle', 'arrow', 'circle', 'square', 'diamond']
print(shape[ :3])
````

- 끝 인덱스의 생략으로 인덱스 -3부터 끝까지 오른쪽으로 이동하여 슬라이싱한다.

````python
shape= ['turtle', 'arrow', 'circle', 'square', 'diamond']
print(shape[-3: ]) 
````

- 시작 인덱스의 생략으로 처음부터 -3 이전까지 오른쪽으로 이동하여 슬라이싱한다.
- 이때 시작의 위치가 인덱스 0이므로 -3 이전은 'arrow'가 된다.

````python
shape= ['turtle', 'arrow', 'circle', 'square', 'diamond']
print(shape[:-3])
````

- 처음부터 -3 이전까지 왼쪽으로 이동하여 슬라이싱한다.
- 이때 시작의 위치가 인덱스 -1이므로 -3 이전은 'square'가 된다.

````python
shape= ['turtle', 'arrow', 'circle', 'square', 'diamond']
print(shape[:-3:-1])
````

- step은 1이고 시작 인덱스, 끝 인덱스의 생략으로 처음부터 끝까지이므로 shape[ : ]와 같다. 

````python
shape= ['turtle', 'arrow', 'circle', 'square', 'diamond']
print(shape[::1])
print(shape[:])
````
