### Problem 1: 저녁 메뉴 추천

😄 1)먹고 싶은 음식 메뉴를 담은 리스트를 하나 만드시오.


> 예 ['샌드위치', '떡볶이', '피자', '햄버거']

2) 앞서 만든 리스트에서 임의로 하나를 택하여, '오늘 저녁 메뉴로 OOO를 추천합니다'와 같은 형식으로 출력하는 코드를 작성하시오

* 임의의 정수는 random.randint 를 사용하여 생성

---
<font color='blue'>`Test case-1`</font>

Output Sample
```
오늘 저녁 메뉴로 샌드위치를 추천합니다.
```
<font color='blue'>`Test case-2`</font>

Output Sample
```
오늘 저녁 메뉴로 피자를 추천합니다.
```

````python
import random

food = ['샌드위치', '떡볶이', '피자', '햄버거']
random_index = random.randint(0, len(food) - 1)
random_food = food[random_index]

print('오늘 저녁 메뉴로 ' + random_food + '를 추천합니다.')
````

### Problem 2: 문자열 개수 파악

😄 문자열을 입력받아 그 문자열에 알파벳 a 가 몇 번 등장하는지를 출력하는 코드를 작성하시오.

---
<font color='blue'>`Test case-1`</font>

Input Sample
```
I wanna dance, the music's got me going ain't nothing that can stop how we move, yeah
```

Output Sample

```
Input:

```

<font color='blue'>`Test case-2`</font>

Input Sample
```
A drowning man will catch at a straw
```

Output Sample

```
Input:

```

````python
text = input('Input: ')
count = 0

for letter in text :
    if (letter == 'a') :
        count = count + 1

print(count)
````

### Problem 3: 동이름 출력하기
😄 구이름과 동이름을 입력 받아 그 중 동이름만 출력하는 프로그램을 작성하시오.

---

Input Sample
```
강동구 노원구 용산구 상암동 서울시 논현동 평창동 중구
```
Output Sample
```
In:
===
동이름
===
상암동
논현동
평창동
```


````python
input_region = input('In: ')

region = input_region.split()

print('=====')
print('동이름')
print('=====')

index = 0

while index < len(region) :
    word = region[index]

    if word[-1] == '동' :
        print(word)
    index = index + 1
````

````python
location = input("구이름과 동이름을 입력하세요").split(' ')

print("=====")
print("동이름")
print("=====")

for dong in location :
  if dong[-1] == "동":
    print(dong)
````

### Problem 4: 중복없이 세 개의 수 출력
😄 1에서 10 까지 수 가운데 중복 없이 세 개의 수를 무작위로 출력하는 코드를 작성하시오. 수는 반드시 오름차순으로 출력되어야 한다.

---

<font color='blue'>`Test case-1`</font>

Output Sample
```
[4, 7, 8, 9]
```

<font color='blue'>`Test case-2`</font>

Output Sample
```
[2, 4, 5, 8]
```

````python
import random

numList = []
while len(numList) <= 3:
  num=random.randint(1,10)
  if num not in numList:
    numList.append(num)
numList.sort()
print(numList)
````

### Problem 6: 369 게임 2
😄 컴퓨터와 사용자 둘 중 한 쪽이 이길 때까지 계속되는 369 게임을 구현하시오. 컴퓨터가
항상 먼저 게임을 시작하며 박수를 두 번 이상 치는 경우는 고려하지 않는다.
컴퓨터는 자신의 차례에서 절대 틀리지 않는다.

---

<font color='blue'>`Test case-1`</font>

Input Sample
```
2
4
짝
8
10
12
짝
```

Output Sample
```
369! 369!
컴퓨터: 1
나:
컴퓨터: 짝
나:
컴퓨터: 5
나:
컴퓨터: 7
나:
컴퓨터: 짝
나:
컴퓨터: 11
나:
컴퓨터: 짝
나:
COMPUTER WIN!
```


````python
print('369! 369!')

num = 1
turn = 0
while True:
  num = str(num)
  if ('3' in num) or ('6' in num) or ('9' in num):
    ans = '짝'
  else:
    ans = num

  if turn == 0:
    print('컴퓨터:', ans)
    turn = 1
  else:
    user_input = input('나: ')
    if user_input == ans:
      turn = 0
    else:
      print('COMPUTER WIN!')
      break
  num = int(num) + 1
````
