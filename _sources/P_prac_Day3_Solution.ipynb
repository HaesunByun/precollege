### Problem 1: 로또 번호 생성기

😄 while문을 이용하여 로또번호 6개를 생성하는 프로그램을 작성하시오.
- 로또번호는 1~45사이 random.randint()를 사용
- 중복 번호를 허용함

---

Output Sample
```
로또번호 생성기 프로그램
1번째 번호는 40입니다.
2번째 번호는 11입니다.
3번째 번호는 11입니다.
4번째 번호는 28입니다.
5번째 번호는 31입니다.
6번째 번호는 5입니다.

산다. 산다. 집 산다!!!
```

````python
import random

print('로또번호 생성기 프로그램')

count = 0
while count < 6 :
    count = count + 1
    lotto_num = random.randint(1, 45)
    print(str(count) + '번째 번호는 ' + str(lotto_num) + '입니다.')

print()
print('산다. 산다. 집 산다!!!')
````

⚠️ 주의: while문의 반복 조건 count < 6에서 < 대신 <=이 쓰이면 어떻게 될지 살펴보도록 하자. 프로그램이 여섯 번째 번호를 출력하고 나서, count 값은 여전히 6이므로 while 문은 멈추지 않게 된다.
이제 count 값이 7로 올라가고 일곱 번째 번호가 출력되고 나서야 while문이 끝날 것이다.
이처럼 반복문은 반복 조건을 어떻게 설정하냐에 따라 결과가 크게 달라질 수 있어 주의가 필요하다.

### Problem 3: 연속 숫자 입력받기
😄 input()으로 사용자로부터 입력받은 자연수를 계속 더해나가다, 음수가 입력되면 그 전까지 계산한 값을 출력하는 코드를 작성하시오. 단, 사용자는 정수만 입력한다

---

Input Sample
```
3
4
10
-5
```
Output Sample
```
Input num:
Input num:
Input num:
Input num:
17
```


````python
total = 0

while True:
    num = int(input('Input num: '))

    if (num < 0) :
        break

    total = total + num

print(total)
````

### Problem 4: 로그인 암호 입력
😄 아래와 같이 올바른 암호를 5회 안에 입력하지 않으면 더 이상 기회를 주지 않도록 프로그램을 작성하시오.

---

<font color='blue'>`Test case-1`</font>

Input Sample
```
pythonis
funfunfun
python
um
cisfun
```
Output Sample
```
암호를 입력하세요:
암호를 입력하세요:
암호를 입력하세요:
암호를 입력하세요:
암호를 입력하세요:
5회 틀렸습니다. 로그인에 실패하셨습니다.
```
<font color='blue'>`Test case-2`</font>

Input Sample
```
funfunfun
예비디딤돌
예비대학
```

Output Sample
```
암호를 입력하세요:
암호를 입력하세요:
암호를 입력하세요:

```

````python
count=0

while count < 5:
    count = count + 1
    pw = input('암호를 입력하세요: ')
    if pw == '예비대학':
        print("You logged in")
        break
else:
    print('5회 틀렸습니다. 로그인에 실패하셨습니다.')
````

📜참고: while True와 같이 반드시 성립하는 반복 조건 을 사용하는 경우 나중에 break 로 반복문을 끝내야 한다.
주어진 문항에서 로그인 시도가 끝나는 경우는 두 가지이다.하나는 로그인 성공 다른 하나는 시도 횟수 초과. 두 경우를 고려하여 반복문을 구상하면 보다 쉽게 코드를 완성할 수 있다.

### Problem 6: 콜라츠 수
😄 독일의 수학자 로타르 콜라츠는 아래와 같은 추측을 제시하였다. 어떤 수가 있을 때, 이 수가 짝수면 2로 나누고 홀수면 3을 곱한 후 1을 더한다.
이 과정을 반복하다 보면 반드시 1 이 나오지 않을까?
콜라츠의 추측을 검증하기 위한 프로그램을 아래 예시와 같이 작성하시오.

---

<font color='blue'>`Test case-1`</font>

Input Sample
```
10
```
Output Sample
```
수를 입력하세요:
5
16
8
4
2
1
역시 콜라츠 대단하네요.
```

````python
num = int(input('수를 입력하세요: '))

while num != 1:
  if (num % 2) == 0:
    num = num // 2
  else:
    num = num * 3 + 1
  print(num)

print('역시 콜라츠 대단하네요.')
````
